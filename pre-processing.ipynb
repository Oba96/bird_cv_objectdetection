```python
import glob
import os
import shutil
```

This are the libraries needed to created new dir and  remove space in the  file(white/blank space)

```python
if not os.path.exists('label/detection/cleaned'):
        os.makedirs('label/detection/cleaned')
else:
    os.system('rm -rf label/detection/cleaned/')
    os.makedirs('label/detection/cleaned')
```

This code will created a dir called dir which will contain the images with their xml file(retriev from the tagging done in VOC pascal)

```python
for file in glob.glob('label/detection/*.xml'):
        print(file)
        with open(file, 'r') as f:
            with open('label/detection/cleaned/%s' %os.path.basename(file), 'w') as f1:
                    for line in f:
                        f1.write(line.rstrip().replace(" ", ""))
                        
```
